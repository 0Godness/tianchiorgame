```python
import os
from pyecharts import Line
```

```python
def show(path):
    d=os.listdir(path)
    d_len=[len(os.listdir(os.path.join(path,i))) for i in d]

    line = Line(path)
    line.add(path, d, d_len, mark_point=["average", "max", "min"])
    return line
```
